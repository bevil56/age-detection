```python
import random
import shutil
import cv2

#Get data
data_dir = "/content/UTKFace"
all_images = [f for f in os.listdir(data_dir) if f.endswith('.jpg.chip.jpg')]


#divide data (80% train, 20% valiation)
random.shuffle(all_images)
train_images = all_images[:int(0.8 * len(all_images))] #80% ƒë·∫ßu
val_images = all_images[int(0.8) *len(all_images):]  #20% cu·ªëi

# Set bounding box sang ƒë·ªãnh d·∫°ng YOLO
def convert_bbox(size, box):
    dw = 1. / size[0]
    dh = 1. / size[1]
    x_center = (box[0] + box[1]) / 2.0
    y_center = (box[2] + box[3]) / 2.0
    width = box[1] - box[0]
    height = box[3] - box[2]
    x_center *= dw
    width *= dw
    y_center *= dh
    height *= dh
    return (x_center, y_center, width, height)

# Move images for train_data
for filename in train_images:
    img_path = os.path.join(data_dir, filename)

    # L∆∞u h√¨nh ·∫£nh t·ª´ dataset ƒë·∫øn file train_images
    shutil.copy(img_path, '/content/yolo_dataset/train/images')

    # Gi·∫£ ƒë·ªãnh r·∫±ng khu√¥n m·∫∑t n·∫±m ·ªü trung t√¢m h√¨nh ·∫£nh v√† chi·∫øm 80% k√≠ch th∆∞·ªõc h√¨nh ·∫£nh
    img = cv2.imread(img_path)
    height, width, _ = img.shape
    box = (0.1 * width, 0.9 * width, 0.1 * height, 0.9 * height)  # Bounding box
    bbox = convert_bbox((width, height), box)

    # T·∫°o t·ªáp nh√£n cho h√¨nh ·∫£nh (train_labels)
    label_path = f'/content/yolo_dataset/train/labels/{os.path.splitext(filename)[0]}.txt'
    with open(label_path, 'w') as f:
        f.write(f'0 {bbox[0]} {bbox[1]} {bbox[2]} {bbox[3]}\n')

# Di chuy·ªÉn h√¨nh ·∫£nh v√† t·∫°o t·ªáp nh√£n cho validation
for filename in val_images:
    img_path = os.path.join(data_dir, filename)

    # L∆∞u h√¨nh ·∫£nh t·ª´ dataset ƒë·∫øn file val_images
    shutil.copy(img_path, '/content/yolo_dataset/val/images')

    # Gi·∫£ ƒë·ªãnh r·∫±ng khu√¥n m·∫∑t n·∫±m ·ªü trung t√¢m h√¨nh ·∫£nh v√† chi·∫øm 80% k√≠ch th∆∞·ªõc h√¨nh ·∫£nh
    img = cv2.imread(img_path)
    height, width, _ = img.shape
    box = (0.1 * width, 0.9 * width, 0.1 * height, 0.9 * height)  # Bounding box
    bbox = convert_bbox((width, height), box)

    # T·∫°o t·ªáp nh√£n cho val_labels
    label_path = f'/content/yolo_dataset/val/labels/{os.path.splitext(filename)[0]}.txt'
    with open(label_path, 'w') as f:
        f.write(f'0 {bbox[0]} {bbox[1]} {bbox[2]} {bbox[3]}\n')

# #di chuy·ªÉn h√¨nh ·∫£nh v√† t·∫°o t·ªáp nh√£n cho train_data
# for filename in train_images:
#   img_path = os.path.join(data_dir, filename)

#   #l∆∞u h√¨nh ·∫£nh t·ª´ dataset ƒë·∫øn file train_images
#   shutil.copy(img_path, '/content/yolo_dataset/train/images')

#   #t·∫°o t·ªáp nh√£n cho h√¨nh ·∫£nh (train_labels)
#   label_path = f'/content/yolo_dataset/train/labels/{os.path.splitext(filename)[0]}.txt'
#   with open(label_path, 'w') as f:
#     f.write('0 0.5 0.5 1.0 1.0\n')


# #di chuy·ªÉn h√¨nh ·∫£nh v√†o t·∫°o t·ªáp nh√£n cho validation
# for filename in val_images:
#   img_path = os.path.join(data_dir, filename)

#   #l∆∞u h√¨nh ·∫£nh t·ª´ dataset ƒë·∫øn file val_images
#   shutil.copy(img_path, '/content/yolo_dataset/val/images')

#   #t·∫°o t·ªáp nh√£n cho val_labels
#   label_path = f'/content/yolo_dataset/val/labels/{os.path.splitext(filename)[0]}.txt'
#   with open(label_path, 'w') as f:
#     f.write('0 0.5 0.5 1.0 1.0\n')


# (0 0.5 0.5 1.0 1.0) : Gi·∫£i ƒë·ªãnh to√†n b·ªô ·∫£nh l√† ƒë·ªëi t∆∞·ª£ng
# Class ID : 0: M√£ s·ªë l·ªõp ƒë·ªëi t∆∞·ª£ng (index : 0) v√¨ ch·ªâ c√≥ 1 class "FACE"
# x_center : 0.5: default bounding box (·ªü gi·ªØa h√¨nh ·∫£nh theo chi·ªÅu ngang)
# y_center : 0.5: default bounding box (·ªü gi·ªØa h√¨nh ·∫£nh theo chi·ªÅu d·ªçc)
# width : 1.0: default bounding box (theo chi·ªÅu ngang)
# height : 1.0: default bounding box (theo chi·ªÅu d·ªçc)



```

#ResNet


```python
# Path to results of Cropping by YOLOv8
output_dir = "/content/cropped_faces_after_YOLO"
os.makedirs(output_dir, exist_ok=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ƒê∆∞·ªùng d·∫´n t·ªõi th∆∞ m·ª•c ch·ª©a d·ªØ li·ªáu UTKFace
data_dir = "/content/UTKFace"
all_images = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.jpg.chip.jpg')]

# Chu·∫©n b·ªã d·ªØ li·ªáu
ages = [int(f.split('_')[0]) for f in os.listdir(data_dir) if f.endswith('.jpg.chip.jpg')]

# T·∫°o DataFrame
df = pd.DataFrame({"image_path": all_images, "age": ages})

# Chia d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm tra
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# T·∫°o ImageDataGenerators
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_dataframe(
    train_df,
    x_col="image_path",
    y_col="age",
    target_size=(224, 224),
    batch_size=32,
    class_mode='raw'
)

validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_dataframe(
    val_df,
    x_col="image_path",
    y_col="age",
    target_size=(224, 224),
    batch_size=32,
    class_mode='raw'
)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model

# Load pretrained ResNet50
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Th√™m c√°c l·ªõp t√πy ch·ªânh cho d·ª± ƒëo√°n tu·ªïi
x = base_model.output
x = Flatten()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='linear')(x)

# ƒê·ªãnh nghƒ©a m√¥ h√¨nh
model = Model(inputs=base_model.input, outputs=predictions)

# ƒê√≥ng bƒÉng c√°c l·ªõp c·ªßa m√¥ h√¨nh c∆° s·ªü ƒë·ªÉ s·ª≠ d·ª•ng c√°c tr·ªçng s·ªë ƒë√£ hu·∫•n luy·ªán tr∆∞·ªõc ƒë√≥
for layer in base_model.layers:
    layer.trainable = False

# Complie model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.summary()

# Hu·∫•n luy·ªán m√¥ h√¨nh v·ªõi d·ªØ li·ªáu UTKFace
history_utkface = model.fit(
    train_generator,
    epochs=2,
    validation_data=validation_generator
)

# L∆∞u m√¥ h√¨nh
model.save("/content/age_prediction_model_utkface_pretrained.h5")
```

#YOLO + ResNet

```python
import cv2
import numpy as np
from ultralytics import YOLO
import os

# Crop images from YOLOv8

# Load YOLO model have trained
yolo_model = YOLO("/content/runs/detect/train3/weights/best.pt")

# folder to contain UTKFace
data_dir = "/content/UTKFace"
output_dir = "/content/cropped_faces_with_YOLO"
os.makedirs(output_dir, exist_ok=True)

# C·∫Øt khu√¥n m·∫∑t v√† l∆∞u l·∫°i
all_images = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.jpg.chip.jpg')]
for img_path in all_images:
    img = cv2.imread(img_path)
    results = yolo_model(img_path)
    for result in results:
        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box)
            face = img[y1:y2, x1:x2]
            face_resized = cv2.resize(face, (224, 224))
            face_filename = os.path.basename(img_path)
            face_path = os.path.join(output_dir, face_filename)
            cv2.imwrite(face_path, face_resized)
```

```python

# ResNet on Yolo

# Chu·∫©n b·ªã d·ªØ li·ªáu
cropped_images = [os.path.join(output_dir, f) for f in os.listdir(output_dir)]
ages = [int(f.split('_')[0]) for f in os.listdir(output_dir)]

# T·∫°o DataFrame
df_cropped = pd.DataFrame({"image_path": cropped_images, "age": ages})

# Chia d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm tra
train_df_cropped, val_df_cropped = train_test_split(df_cropped, test_size=0.2, random_state=42)

# T·∫°o ImageDataGenerators
train_generator_cropped = train_datagen.flow_from_dataframe(
    train_df_cropped,
    x_col="image_path",
    y_col="age",
    target_size=(224, 224),
    batch_size=32,
    class_mode='raw'
)

validation_generator_cropped = validation_datagen.flow_from_dataframe(
    val_df_cropped,
    x_col="image_path",
    y_col="age",
    target_size=(224, 224),
    batch_size=32,
    class_mode='raw'
)

# Hu·∫•n luy·ªán m√¥ h√¨nh v·ªõi d·ªØ li·ªáu khu√¥n m·∫∑t ƒë√£ c·∫Øt t·ª´ YOLO
history_yolo_resnet = model.fit(
    train_generator_cropped,
    epochs=2,
    validation_data=validation_generator_cropped
)

# L∆∞u m√¥ h√¨nh
model.save("/content/age_prediction_model_yolo_resnet.h5")
```

```python
import matplotlib.pyplot as plt

# V·∫Ω ƒë·ªì th·ªã loss
plt.figure(figsize=(12, 8))
plt.plot(history_utkface.history['loss'], label='Train Loss (UTKFace Pretrained)')
plt.plot(history_utkface.history['val_loss'], label='Validation Loss (UTKFace Pretrained)')
plt.plot(history_yolo_resnet.history['loss'], label='Train Loss (YOLO + ResNet)')
plt.plot(history_yolo_resnet.history['val_loss'], label='Validation Loss (YOLO + ResNet)')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# V·∫Ω ƒë·ªì th·ªã MAE
plt.figure(figsize=(12, 8))
plt.plot(history_utkface.history['mae'], label='Train MAE (UTKFace Pretrained)')
plt.plot(history_utkface.history['val_mae'], label='Validation MAE (UTKFace Pretrained)')
plt.plot(history_yolo_resnet.history['mae'], label='Train MAE (YOLO + ResNet)')
plt.plot(history_yolo_resnet.history['val_mae'], label='Validation MAE (YOLO + ResNet)')
plt.title('Model Mean Absolute Error')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()
```
